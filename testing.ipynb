Data used in this project is The Ciqual table and it was obtained through the CIQUAL project website: [ciqual.anses.fr](https://ciqual.anses.fr/#/cms/download/node/20)
<br>
<br>
The Ciqual table is produced by the Food Observatory unit of the <b>French Agency for Food, Environmental and Occupational Health & Safety (Anses)</b>.
<br>
<br>
This dataset provides the nutritional composition of over <b>3,484</b> foods consumed in France: lipids, proteins, carbohydrates, sugars, fatty acids, salt, vitamins and minerals. It is one of the most comprehensive food composition tables in Europe.
<br>
<br>
The purpose of this dataset is to be a reference tool for various uses:
<br>
<br>
Manufacturers for labelling purposes
<br>
Health professionals, dieticians and doctors, to monitor their patients
<br>
Researchers in nutritional epidemiology
<br>
More generally, anyone with a particular interest in food composition
<br>
<br>
The Ciqual table is also used by Anses scientists to calculate nutritional intakes in the framework of food consumption surveys, and for nutritional risk assessment.
<br>
<br>
Obtained on 15/05/2026 
<br>
